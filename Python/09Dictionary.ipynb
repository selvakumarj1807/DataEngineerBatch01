# Dictionary

```python
studentData = {'studId': 101, 'studName': 'Sara', 'age': 25}

print(studentData, type(studentData))
```

```python
print(studentData['studName'])
```

# Update

```python
studentData = {'studId': 101, 'studName': 'Sara', 'age': 25}

print(studentData)

print('----------------------------------------')

studentData.update({'gender': 'male', 'city': 'Chennai'})

print(studentData)
```

# Pop

```python
studentData = {'studId': 101, 'studName': 'Sara', 'age': 25}

print(studentData)

print('----------------------------------------')

studentData.pop('age')

print(studentData)
```

```python
studentData = {'studId': 101, 'studName': 'Sara', 'age': 25}

print(studentData)

print('----------------------------------------')

print(studentData.keys())

studentList = studentData.values()

print(studentList)

print(studentData.items())
```

# For Loop

```python
studentData = {'studId': 101, 'studName': 'Sara', 'age': 25}

print(studentData)

print('----------------------------------------')

for i in studentData:
    print(i, ' - ', studentData[i])
```

```python
for i in range(10):
    print(i)
```
