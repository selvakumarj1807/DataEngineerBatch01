# Set

```python
studentData = {101, 'Sara', 'Female', 'Sara', 'Female', 'Sara', 'Female'}

print(studentData, type(studentData))
```

# Add

```python
studentData = {101, 'Sara', 'Female'}

print(studentData, type(studentData))

print('------------------------------------')

studentData.add(25)

print(studentData)
```

# Update

```python
studentData = {101, 'Sara', 'Female'}

print(studentData, type(studentData))

print('------------------------------------')

data02 = {25, 'CHE', 'TN'}

studentData.update(data02)

print(studentData)
```

# Remove

```python
studentData = {101, 'Sara', 'Female'}

print(studentData, type(studentData))

print('------------------------------------')

studentData.remove('Sara')

print(studentData)
```

# Discard

```python
studentData = {101, 'Sara', 'Female'}

print(studentData, type(studentData))

print('------------------------------------')

studentData.discard('Sara')

print(studentData)
```

# Diff between Remove & Discard

```python
studentData = {101, 'Sara', 'Female'}

print(studentData, type(studentData))

print('------------------------------------')

studentData.discard('CHE')

print(studentData)
```

```python
studentData = {101, 'Sara', 'Female'}

print(studentData, type(studentData))

print('------------------------------------')

studentData.remove('CHE')

print(studentData)
```

# Set Operators

# union

```python
data01 = {1, 2, 3, 4, 5, 7}

data02 = {1, 2, 3, 6, 8, 9, 10}

r1 = data01.union(data02)

print(r1)
```

# Intersection

```python
data01 = {1, 2, 3, 4, 5, 7}

data02 = {1, 2, 3, 6, 8, 9, 10}

r2 = data01.intersection(data02)

print(r2)
```

# Symmetric_difference

```python
data01 = {1, 2, 3, 4, 5, 7}

data02 = {1, 2, 3, 6, 8, 9, 10}

r2 = data01.symmetric_difference(data02)

print(r2)
```
