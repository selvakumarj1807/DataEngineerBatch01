# Type Casting

```python
a = 10

b = 90.5

print(a, type(a))

print(b, type(b))

print('---------------------------------')

a = float(a)

b = int(b)

print(a, type(a))

print(b, type(b))
```

```python
age = '25'

print(age, type(age))

birthYear = 2026 - int(age)

print(birthYear)
```

# The area of a circle is calculated using the formula \(A = \pi r^2\)

```python
pi = 3.14

r = 10

A = pi * r * r

print('The area of a circle (', r, ') is : ', A)

print(f'The area of a circle({r}) is : {A}')
```
