# Upper, Lower, Title, Capitalize

```python
msg = 'wELcome PyTHoN proGrammING laNguage'

print(msg.upper())

print(msg.lower())

print(msg.title())

print(msg.capitalize())
```

```python
products = ['Mobile', 'Laptop', 'MensCloth', 'WomensCloth']

searchproduct = input('Enter your Product : ')

print(searchproduct.title())

print(searchproduct.title() in products)
```

# isupper, islower, isalpha, isalnum

```python
name = 'RAM'

print(name.isupper())

name = 'ram'

print(name.islower())

name = 'RamPrasanth'

print(name.isalpha())

name = 'ram123'

print(name.isalnum())
```

# len, strip, lstrip, rstrip

```python
mobile = '  9500  '

print(len(mobile))

print (len(mobile.strip()))

print('---------------------------')

mobile = '##9500##'

print(mobile.strip('#'))

print('---------------------------')

mobile = '##9500##'

print(mobile.lstrip('#'))

print('---------------------------')

mobile = '##9500##'

print(mobile.rstrip('#'))
```

# find, count

```python
msg = 'welcome python programming language'

print(msg.count('o'))

print(msg.find('o'))

print(msg.find('o', 8)) # find(findvalue, start index position)
```

# Split

```python
userData = '101, Sara, 22, sara@gmail.com, Chennai'

print(userData, type(userData))

listUserData = userData.split(', ')

print(listUserData, type(listUserData))
```

# Split Lines

```python
courses = '''The three battalions serving with the Parachute Brigade were removed from their respective 
Infantry Regiments to form the Parachute Regiment. Since then the Parachute Regiment has grown to comprise ten battalions 
including Parachute (Special Forces) battalions. 
In 1986, 8 PARA became 12 Battalion, Mechanised Infantry Regiment, 
while 21 Maratha LI converted to PARA (Special Forces).  '''

print(courses)

print('-----------------------------------------------')

splitPara = courses.splitlines()

print(splitPara)
```

# Joins

```python
listStudent = ['101', 'Sara', '22', 'sara@gmail.com', 'Chennai']

joinStudent = ', '.join(listStudent)

print(joinStudent, type(joinStudent))
```

```python
date = ['29', '6', '2026']

print('/'.join(date))
```

# String Manipulation

```python
message = "Welcome to Python"

print(message, type(message))

print('---------------------------------')

print(message[2])

print(message[8:9 + 1])

print(message[:7])

print(message[11:])

print(message[-6:])

print(message[-9:-8 + 1])

print(message[-9:-7])
```
