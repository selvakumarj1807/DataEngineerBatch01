# List

```python
studentData = [101, 'Sara', 'Female']

print(studentData)
```

# Change Values Using Index Position

```python
print(studentData[1])

print('------------------------------')

studentData[1] = 'Priya'

print(studentData)
```

# Append

```python
studentData = [101, 'Sara', 'Female']

print(studentData)

print('------------------------------')

studentData.append(25)

print(studentData)
```

# Extend

```python
studentData = [101, 'Sara', 'Female']

print(studentData)

print('------------------------------')

data01 = [25, 9500912258, 'CHE', 'TN']

studentData.extend(data01)

print(studentData)
```

# Insert

```python
studentData = [101, 'Sara', 'Female']

print(studentData)

print('------------------------------')

studentData.insert(1, 'student')

print(studentData)
```

# Pop - Remove Values using Index Position

```python
studentData = [101, 'Sara', 'Female']

print(studentData)

print('------------------------------')

studentData.pop(1)

print(studentData)
```

# Remove - Remove Values using give Value

```python
studentData = [101, 'Sara', 'Female', 'Sara']

print(studentData)

print('------------------------------')

studentData.remove('Sara')

print(studentData)
```

```python
age = [10, 20, 10, 28, 22, 50, 20]

print(sum(age))

print(min(age))

print(max(age))

print(len(age))

print(age.count(20))
```
