# Python User Defined Functions

# 01 No Return Type With Out Argument

```python
# Create the Function 

def welcome():
    print('Welcome Python!...')
```

```python
# Call the Function

welcome()

welcome()

welcome()
```

```python
# Create the Function

def addTwoNumber():
    a = int(input('Enter value 01 : '))

    b = int(input('Enter value 02 : '))

    print(f"{a} + {b} = {a + b}")
```

```python
# Call the Function

addTwoNumber()

print('------------------------')

addTwoNumber()
```

# 02 No Return Type With Argument

```python
# Create the Function

def addTwoNumber(a, b):
    print(f"{a} + {b} = {a + b}")
```

```python
# Call the Function

addTwoNumber(10, 2)

print('------------------------')

addTwoNumber(20, 2)
```

# 03 Return Type With Out Argument

```python
# Create the Function

def addTwoNumber():
    a = int(input('Enter value 01 : '))

    b = int(input('Enter value 02 : '))

    c = a + b

    return c
```

```python
# Call the Function

print('From Print Statement : ', addTwoNumber())

print('----------------------------------')

r1 = addTwoNumber()

print('From Variable Declaration : ', r1)
```

# 04. Return Type With Argument

```python
# Create the Function

def addTwoNumber(a, b):
    c = a + b

    return c
```

```python
# Call the Function

print('From Print Statement : ', addTwoNumber(10, 2))

print('----------------------------------')

r1 = addTwoNumber(20, 2)

print('From Variable Declaration : ', r1)
```

# Arbitrary Argument Function

```python
# Create the Function

def userData(*data):
    print(data, type(data))
```

```python
# Call the Function

userData(101, 'Ram', 27, 9500912258)
```

# Keyword Argument Function

```python
# Create the Function

def userData(userid, username, age):
    print(userid, ' : ', ' Hi ', username, ' your age is ', age)
```

```python
# Call the Function

userData(101, 'Ram', 27)

userData(userid=102, age=22, username='Sam')
```

# Keyword Arbitrary Argument Function

```python
# Create the Function

def userData(**data):
    print(data)
```

```python
# Call the Function

userData(userid=102, age=22, username='Sam')
```

# Default Parameter Function

```python
# Create the Function

def userData(userid, username, age, city='TamilNadu'):
    print(userid, ' : ', ' Hi ', username, ' your age is ', age, ' your city is ', city)
```

```python
# Call the Function

userData(101, 'Ram', 27, 'Kerala')

userData(userid=102, age=22, username='Sam')
```

# Pass the List values

```python
# Create the Function

def userData(data):
    print(data, type(data))
```

```python
# Call the Function

userData([101, 'Ram', 27, 9500912258])
```

# Lambda Function

```python
addTwoNumber = lambda a,b: f"{a} + {b} = {a + b}"

print(addTwoNumber(10, 2))

print(addTwoNumber(20, 2))
```
