# Tuple

```python
studentData = (101, 'Sara', 'Female')

print(studentData, type(studentData))
```

```python
print(studentData[1])

print('----------------------------------')

studentData = list(studentData)

studentData[1] = 'Priya'

print(studentData, type(studentData))

print('----------------------------------')

studentData = tuple(studentData)

print(studentData, type(studentData))
```

# For

```python
studentData = (101, 'Sara', 'Female')

print(studentData, type(studentData))

print('----------------------------------')

for i in studentData:
    print(i)
```

# If

```python
studentData = (101, 'Sara', 'Female')

print(studentData, type(studentData))

print('----------------------------------')

if 'Sara' in studentData:
    print('Yes')
else:
    print('No')
```
