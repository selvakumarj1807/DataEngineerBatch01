# IF

```python
age = 26

if age >= 18:
    print(age, ' is Above 18. So you are eligible for vote')
```

# If & Else

```python
age = 12

if age >= 18:
    print(age, ' is Above 18. So you are eligible for vote')

else:
    print(age, ' is Below 18. So you are not eligible for vote')
```

# If & Elif & Else

```python
value = -12

if value > 0:
    print(value, ' is +ve Value')

elif value == 0:
    print(value, ' is 0')

elif value == -1:
    print(value, ' is 1st Nagative Value')

else:
    print(value, ' is -ve Value')
```

# Using Logical Operators

```python
userName = input('Enter your User Name : ')

password = input('Enter your Password : ')

print('------------------------------')

if ((userName == 'admin') and (password == 'Admin@123')):
    print('Welcome Admin!...')

else:
    print('Invalid Credentials!...')
```

# Nested IF

```python
m1 = 23

m2 = 90

m3 = 22

if m1 >= 35 and m2 >= 35 and m3 >= 35:
    total = m1 + m2 + m3

    avg = int(total / 3)
    
    print('Total : ', total)
    print('Average : ', avg)

    if avg >= 90 and avg <= 100:
        print(avg, ' - A Grade')
        
    elif avg >= 80 and avg <= 90:
        print(avg, ' - B Grade')

    else:
        print(avg, ' - No Grade')

else:
    print('You are Under 35')

    if m1 <= 35:
        print(m1, ' - Mark 01 is Below 35')
    else:
        print(m1, ' - Mark 01 is Above 35')

    if m2 <= 35:
        print(m2 , ' - Mark 02 is Below 35')
    else:
        print(m2, ' - Mark 02 is Above 35')

    if m3 <= 35:
        print(m3, ' - Mark 03 is Below 35')
    else:
        print(m3, ' - Mark 03 is Above 35')
```
