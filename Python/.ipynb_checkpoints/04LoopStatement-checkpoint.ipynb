# While Loop

```python
i = 0 # Value Initial

while i <= 5: # 6 <= 5 Value Condition
    print(i)

    i = i + 1 # Value Increment
```

# To Generate Even Numbers

```python
i = 0 

while i <= 10: # 12 <= 10
    print(i)

    i = i + 2 
```

# To Gerate Odd Numbers

```python
i = 1 # Value Intial 

while i <= 10: # Value Condition
    print(i)

    i = i + 2 # Value Increment
```

# Jumping Statements 

# 1. Break

```python
i = 0 

while i <= 10: 
    print(i)

    if i == 5:
        print('Before Break Statement')
        i = i + 1
        break
        print('After Break Statement')

    i = i + 1 
```

# 2. Continue

```python
i = 0 

while i <= 10: 
    print(i)

    if i == 5:
        print('Before Continue Statement')
        i = i + 2
        continue
        print('After Continue Statement')

    i = i + 1 
```

# 3. Pass

```python
i = 0 

while i <= 10: 
    print(i)

    if i == 5:
        print('Before Pass Statement')
        i = i + 2
        pass
        print('After Pass Statement')

    i = i + 1 
```

# For Loop

```python
# Default Value Initial 0
# Default Value Increment 1

for i in range(10):
    print(i)
```

# To Generate Odd Numbers

```python
for i in range(1, 10, 2):
    print(i)
```

```python
for i in range(1, 10, 2):
    print(i, end=', ')
```

# To Generate Even Numbers

```python
for i in range(0, 10, 2):
    print(i)
```

# Boolean Values Using Loop Condition

```python
condition = True

i = 0

while condition:
    print('Welcome Python!....')

    if i == 2:
        condition = False

    i += 1 # i = i + 1        
```

# Table

```python
limit = 5

table = 2

i = 1

while i <= limit:
    print(i, ' X ', table, ' = ', i * table)

    i += 1
```
