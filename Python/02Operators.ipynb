# 1. Arithmatic Operators

```python
a = 7

b = 2

print('Addition (+) : ', a + b)

print('Subtraction (-) : ', a - b)

print('Multiplication (*) : ', a * b)

print('Divsion (/) : ', a / b)

print('-------------------------------------')

print('Modulas (%) :', a % b)

print('Floor Divsion (//) : ', a // b)

print('Exponentiation (**) : ', a ** 2) # 7 * 7
```

# 2. Arithmatic Assignment Operators

```python
# Assignment Operators

a = 10

print('Assignment Operator (=): ', a)

# Arithmatic Assignment Operators

a += 1 # a = a + 1

print('Arithmatic Assignment Operators (+=) : ', a)
```

# 3. Comparision Operators

```python
a = 10

b = 2

print('Equal : ', a == b)

print('Not Equal :', a != b)

print('Grater then : ', a > b) # 10 > 2

print('Less then : ', a < b) # 10 < 2

print('Grater then or Equal : ', a >= 10) # 10 > 10

print('Less then or Equal : ', a <= 10) # 10 < 10
```

# 4. Logical Operators

```python
username = input('Enter your User Name : ')

password = input('Enter your Password : ')

print(username == 'admin')

print(password == 'admin123')

print('--------------------------------------')

# Logical And Operators
'''
1. Every condition have True. then always True
2. Any one condition have False. Then everything is False
'''

print('Logical And Operators : ', (username == 'admin') and (password == 'admin123'))

print('--------------------------------------')

# Logical Or Operators
'''
1. Any one condition have True. Then everything is True
'''

print('Logical Or Operators : ', (username == 'admin') or (password == 'admin123'))

print('--------------------------------------')

# Not Operators
'''
True => False
False => True
'''

print(not(username == 'admin'))

print('not(Logical And Operators) : ', not ((username == 'admin') and (password == 'admin123')))
```

# 5. Membership Operators

```python
ageValues = [10, 'Sara', 95]
             
print('Lara' in ageValues)

print('Lara' not in ageValues)
```

# 6. Identity Operators

```python
a = 10

b = 10

print(id(a))

print(id(b))

print('-------------------------')

print(a is b)

print(a is not b)
```

# 7. Bitwise Operators

```python
'''
a = 10 = 1010

b = 2  = 0010 (&). 0 = False, 1 = True. Any one is False. Then everything is False
--------------
         0010
'''

a = 10

b = 2

print('Bitwise and Operator (&) : ', a & b)
```

```python
'''
a = 10 = 1010

b = 2  = 0010 (&). 0 = False, 1 = True. Any one is True. Then everything is True
--------------
         1010
'''

a = 10

b = 2

print('Bitwise or Operator (|) : ', a | b)
```

```python
'''
a = 10 = 1010

b = 2  = 0010 (&). 0 = False, 1 = True. Both are Same(True or False). Then return False
--------------
         1000
'''

a = 10

b = 2

print('Bitwise Xor Operator (^) : ', a ^ b)
```

```python
'''
a = 10
~a = -(a + 1)
   = -(10 + 1)
   = -11
'''

a = 10

print('Bitwise Not Operator (~) : ', ~a)
```

```python
'''
a << 2 = 10 << 2
       = 10 * (2 * 2)
       = 40

a << 3 = 10 << 3
       = 10 * (2 * 2 * 2)
       = 80
'''

a = 10

print('Left shift Operator(<<) : ', a << 2)

print('Left shift Operator(<<) : ', a << 3)
```

```python
'''
a >> 2 = 10 >> 2
       = 10 / (2 * 2)
       = 10 / 4
       = 40

a >> 3 = 10 >> 3
       = 10 / (2 * 2 * 2)
       = 10 / 8
       = 80
'''

a = 10

print('Right shift Operator(>>) : ', a >> 2)

print('Right shift Operator(>>) : ', a >> 3)
```
