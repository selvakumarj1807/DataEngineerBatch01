# List

1. List is a Collection of Data Type
2. It can declare [ ]. Assign values Index Type
3. List are Mutable(can change or alter values)

```python
studentData = [101, 'Ram', 25, 'Male', 9500912258]

print(studentData, type(studentData))

print('--------------------------------------')

print(studentData[0])

print(studentData[1])

studentData[1] = 'Ram Kumar'

print(studentData)
```

# Tuple

1. Tuple is a Collection of Data Type
2. It can declare ( ). Assign values Index Type
3. Tuple are ImMutable(can not change or alter values)

```python
studentData = (101, 'Ram', 25, 'Male', 9500912258)

print(studentData, type(studentData))

print('--------------------------------------')

# TypeError: 'tuple' object does not support item assignment

print(studentData[0])

print(studentData[1])

studentData[1] = 'Ram Kumar'

print(studentData)
```

# 3. Set

1. Set is a Collection of Data Type
2. It can declare { }. Assign values Non Index Type
3. Set are Mutable(can change or alter values)
4. Remove Douplicate Values

```python
ageValues = {20, 10, 22, 10, 20, 20}

print(ageValues, type(ageValues))
```

# Dictionary

1. Dictionary is a Collection of Data Type
2. It can declare {'key': 'value'}. Assign values Non Index Type
3. Dictionary are Mutable(can change or alter values)
4. Remove Douplicate Values

```python
studentData = {'studID': 101, 'studName': 'Ram', 'age': 25, 'gender': 'Male', 'mobile': 9500912258}

print(studentData, type(studentData))

print('---------------------------------------------------')

print(studentData['studID'])
```
